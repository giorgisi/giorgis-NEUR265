```python
#Imports relevant packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Import data sets
urllt1 = 'https://raw.githubusercontent.com/giorgisi/giorgis-NEUR265/main/spikes_left_trial1.csv'
urlrt1 = 'https://raw.githubusercontent.com/giorgisi/giorgis-NEUR265/main/spikes_right_trial1.csv'
urllt2 = 'https://raw.githubusercontent.com/giorgisi/giorgis-NEUR265/main/spikes_left_trial2.csv'
urlrt2 = 'https://raw.githubusercontent.com/giorgisi/giorgis-NEUR265/main/spikes_right_trial2.csv'

spikes_left_trial1 = np.loadtxt(urllt1, delimiter=',')
spikes_right_trial1 = np.loadtxt(urlrt1, delimiter=',')
spikes_left_trial2 = np.loadtxt(urllt2, delimiter=',')
spikes_right_trial2 = np.loadtxt(urlrt2, delimiter=',')
```

```python
#Prints the length of each array

print('spikes_left_trial1:', len(spikes_left_trial1))
print('spikes_right_trial1:', len(spikes_right_trial1))
print('spikes_left_trial2:', len(spikes_left_trial2))
print('spikes_right_trial2:', len(spikes_right_trial2))
```

```python
#Determine firing rate of each neuron
time = 30

print('Left Trial 1:', len(spikes_left_trial1)/time, 'Hz')
print('Right Trial 1:', len(spikes_right_trial1)/time, 'Hz')
print('Left Trial 2:', len(spikes_left_trial2)/time, 'Hz')
print('Right Trial 2:', len(spikes_right_trial2)/time, 'Hz')
```

* There doesn't seem to be a distinct difference in firing frequency between left and right trials. All values were between 1.5-2.3 Hz.


* There also doesn't seem to be a trend between trials (trial 1 values vs trial 2 values)

```python
# Make subplots to graph the spike train data

fig, ax = plt.subplots(1,2,figsize=(15,5))

ax[0].plot(spikes_left_trial1, label='Left')
ax[0].plot(spikes_right_trial1, label='Right')
ax[0].set_title('Trial 1')
ax[0].set_xlabel('Index')
ax[0].set_ylabel('Voltage [$\mu V$]')
ax[0].legend()

ax[1].plot(spikes_left_trial2, label='Left')
ax[1].plot(spikes_right_trial2, label='Left')
ax[1].set_title('Trial 2')
ax[1].set_xlabel('Index')
ax[1].set_ylabel('Voltage [$\mu V$]')
ax[0].legend()
```

* In the first trial, the right neurons seems to have a very large burst, while the left neurons seems to be more steady

* In the second trial, the left neruon has a burst while the right is more steady



```python
# Calculate ISI for each of the spike arrays

spikes_left_trial1_ISI = np.diff(spikes_left_trial1)
spikes_right_trial1_ISI = np.diff(spikes_right_trial1)

spikes_left_trial2_ISI = np.diff(spikes_left_trial2)
spikes_right_trial2_ISI = np.diff(spikes_right_trial2)
```

```python
# Make histograms for each ISI array

bins = np.arange(0,2,0.01)

fig, ax = plt.subplots(2,2,figsize=(15,10))
ax[0,0].hist(spikes_left_trial1_ISI, bins)
ax[0,0].set_title('Left Trial 1 ISI')
ax[0,0].set_xlabel('ISI')
ax[0,0].set_ylabel('Count')

ax[0,1].hist(spikes_right_trial1_ISI, bins)
ax[0,1].set_title('Right Trial 1 ISI')
ax[0,1].set_xlabel('ISI')
ax[0,1].set_ylabel('Count')

ax[1,0].hist(spikes_left_trial1_ISI, bins)
ax[1,0].set_title('Left Trial 2 ISI')
ax[1,0].set_xlabel('ISI')
ax[1,0].set_ylabel('Count')

ax[1,1].hist(spikes_left_trial1_ISI, bins)
ax[1,1].set_title('Right Trial 2 ISI')
ax[1,1].set_xlabel('ISI')
ax[1,1].set_ylabel('Count')
```

* The left trial seem to have slightly more range of ISI, compared to the right trials
* All four data sets show bursting behavior, because there is a signifcant range of different ISIs, indicating they fire irregularly
* Right trial 1 has the most spread of ISIs

```python
# Calculate the firing rates for each array, plot

binsize = 1
fr_bins = np.arange(0,30,binsize)
fr_leftt1 = np.zeros(len(fr_bins)-1)

for i in range(len(fr_bins)-1):
  spk_index = (spikes_left_trial1 > fr_bins[i]) & (spikes_left_trial1 < fr_bins[i+1])
  true_values = spikes_left_trial1[spk_index]
  fr_leftt1[i] = (len(true_values))/binsize

fr_rightt1 = np.zeros(len(fr_bins)-1)
for i in range(len(fr_bins)-1):
  spk_index = (spikes_right_trial1 > fr_bins[i]) & (spikes_right_trial1 < fr_bins[i+1])
  true_values = spikes_right_trial1[spk_index]
  fr_rightt1[i] = (len(true_values))/binsize

fr_leftt2 = np.zeros(len(fr_bins)-1)
for i in range(len(fr_bins)-1):
  spk_index = (spikes_left_trial2 > fr_bins[i]) & (spikes_left_trial2 < fr_bins[i+1])
  true_values = spikes_left_trial2[spk_index]
  fr_leftt2[i] = (len(true_values))/binsize

fr_rightt2 = np.zeros(len(fr_bins)-1)
for i in range(len(fr_bins)-1):
  spk_index = (spikes_right_trial2 > fr_bins[i]) & (spikes_right_trial2 < fr_bins[i+1])
  true_values = spikes_right_trial2[spk_index]
  fr_rightt2[i] = (len(true_values))/binsize


fig, ax = plt.subplots(1,2,figsize=(15,10))
ax[0].plot(fr_bins[:-1], fr_leftt1, 'r', label ='Left')
ax[0].plot(fr_bins[:-1], fr_rightt1, 'blue', label ='Right')
ax[0].set_title('Trial 1')
ax[0].set_xlabel('Time [s]')
ax[0].set_ylabel('Firing Rate [Hz]')
ax[0].legend()

ax[1].plot(fr_bins[:-1], fr_leftt2, 'r', label ='Left')
ax[1].plot(fr_bins[:-1], fr_rightt2, 'blue', label ='Right')
ax[1].set_title('Trial 2')
ax[1].set_xlabel('Time [s]')
ax[1].set_ylabel('Firing Rate [Hz]')
ax[1].legend()
```

* In Trial 1, the left neuron peaks around 15 seconds, with two large spikes. The right trial has less significant peaks, around 6 and 9 seconds.


* In Trial 2, the right neuron peaks very eary, around 3 seconds, and then twice more at 10 and 15s. The left neuron is more steady in this trial, with one smaller peak around 4s and a series of peaks from 20-25s.
