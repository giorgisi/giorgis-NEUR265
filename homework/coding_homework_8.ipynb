```python
#Import relevant modules

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

```

```python
#Import data sets

fear_traces = np.loadtxt('https://raw.githubusercontent.com/giorgisi/giorgis-NEUR265/main/fear_conditioning_traces_hw8.csv', delimiter=',')
extinction_predictors = np.loadtxt('https://raw.githubusercontent.com/giorgisi/giorgis-NEUR265/main/extinction_predictors_hw8.csv', delimiter=',')
shock_index = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265/main/data/shock_index.csv', delimiter=',')
tone_index = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265/main/data/tone_index.csv', delimiter=',')
```

```python
#Create conditioning and extinction variables

conditioning_traces = fear_traces[0:10457,:]
extinction_traces = fear_traces[10458:,:]
```

```python
 #Print size of arrays

 print(np.shape(conditioning_traces))
 print(np.shape(extinction_traces))
```

```python
#Create Time Variables
conditioning_intervals = len(conditioning_traces)/15
extinction_intervals = len(extinction_traces)/15

print(conditioning_intervals, extinction_intervals)

time_conditioning = np.linspace(0,697.1,10457)
time_extinction = np.linspace(0,947.3,14210)

print(len(time_conditioning), len(time_extinction))
```

* The conditioning session lasted about 697.1 seconds, or 11.6 minutes
* The extinction session lasted about 947.3 seconds, or 15.8 minutes
* 63 cells were imaged during these sessions


```python
#Plot the first and second ROI from conditioning, first and second ROI for extinction in subplots
fig, ax = plt.subplots(1,2,figsize=(15,5))

ax[0].plot(time_conditioning, conditioning_traces[:,0], 'b', alpha=0.5)
ax[0].plot(time_conditioning, conditioning_traces[:,1], 'r', alpha=0.5)
ax[0].set_xlabel('Time [s]')
ax[0].set_ylabel('Voltage [$\mu V$]')
ax[0].set_title('Conditioning')

ax[0].plot([shock_index[0], shock_index[0]], [0,1], 'k', lw=1)
ax[0].plot([shock_index[1], shock_index[1]], [0,1], 'k', lw=1)
ax[0].plot([shock_index[2], shock_index[2]], [0,1], 'k', lw=1)
ax[0].plot([tone_index[0], tone_index[0]], [0,1], 'g', lw=1)
ax[0].plot([tone_index[1], tone_index[1]], [0,1], 'g', lw=1)
ax[0].plot([tone_index[2], tone_index[2]], [0,1], 'g', lw=1)

ax[1].plot(time_extinction, extinction_traces[:,0], 'b', alpha=0.5)
ax[1].plot(time_extinction, extinction_traces[:,1], 'r', alpha=0.5)
ax[1].set_xlabel('Time [s]')
ax[1].set_ylabel('Voltage [$\mu V$]')
ax[1].set_title('Extinction')


```

* For the first and second ROIs in conditioning, there is a significant spike after shock onset, but the spike is slightly smaller each time
* The first ROI peaks higher, but in general the second ROI is slightly higher

* The extinction has less significant peaks for both, except for the second ROI which has a large peak right around 500s

```python
#Make predictors_conitioning variable
shock_array = np.zeros(10457)
for i in range(10457):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(10457)
for i in range(10457):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0

from pandas import DataFrame as df
import statsmodels.api as sm
from statsmodels.genmod.families import Poisson
from statsmodels.genmod.families.links import identity, log
from scipy.stats import chi2
from statsmodels.distributions.empirical_distribution import ECDF

first_trace = conditioning_traces[:,0]
predictors_conditioning = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array})

predictors_conditioning['shock onset'] = shock_array

#print(np.shape(predictors_conditioning), np.shape(conditioning_traces))
```

```python
# Run model for each ROI for each

predictors_conditioning['shock onset'] = shock_array
r_model_conditioning = np.zeros(63)

for i in range(63):
  trace_temp = conditioning_traces[:,i]
  model3 = sm.GLM(trace_temp, predictors_conditioning, family=Poisson())
  model3_results = model3.fit()
  predicted_trace = model3_results.predict()
  r_index_conditioning = pearsonr(trace_temp, predicted_trace)
  r_model_conditioning[i] = r_index_conditioning.statistic
```

```python
# Run model for each ROI for extinction

shock_array = np.zeros(14210)
for i in range(14210):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(14210)
for i in range(14210):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0

from pandas import DataFrame as df
import statsmodels.api as sm
from statsmodels.genmod.families import Poisson
from statsmodels.genmod.families.links import identity, log
from scipy.stats import chi2
from statsmodels.distributions.empirical_distribution import ECDF

first_trace = extinction_traces[:,0]
predictors_extinction = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array})

predictors_extinction['shock onset'] = shock_array
r_model_extinction = np.zeros(63)

for i in range(63):
  trace_temp = extinction_traces[:,i]
  model3 = sm.GLM(trace_temp, predictors_extinction, family=Poisson())
  model3_results = model3.fit()
  predicted_trace = model3_results.predict()
  r_index_extinction = pearsonr(trace_temp, predicted_trace)
  r_model_extinction[i] = r_index_extinction.statistic
```

```python
# Use pearsonr function to correlate these two

correlation = pearsonr(r_model_conditioning, r_model_extinction)
correlation
```

```python
# Make a scatterplot with the two models
plt.scatter(r_model_conditioning, r_model_extinction)
plt.xlabel('r_model_conditioning')
plt.ylabel('r_model_extinction')
```

There does appear to be a correlation between the extinction and condition cells - as response to conditioning increases, response to extinction also increase. You could likely draw a linear correlation here, but there is still a significant amount of noise in the data.
