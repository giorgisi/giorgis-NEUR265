```python
#Imports nump, matplotlib.pyplot, and scipy.stats

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#Imports v_rest data as numpy arrays

v_rest1_url = 'https://raw.githubusercontent.com/giorgisi/giorgis-NEUR265/main/v_rest_sst.csv'
v_rest2_url = 'https://raw.githubusercontent.com/giorgisi/giorgis-NEUR265/main/v_rest_pvalb.csv'

v_rest_sst = np.loadtxt(v_rest1_url)
v_rest_pvalb = np.loadtxt(v_rest2_url)
```

```python
#Finds means of v_rest_sst and v_rest_pvalb
print(np.mean(v_rest_sst))
print(np.mean(v_rest_pvalb))

#t-test to find if means are different
t_sat, pval = (stats.ttest_ind(v_rest_sst, v_rest_pvalb))
print(pval)

#Plot histogram of v_rest_sst and v_rest_pvalb with subplots if p > 0.05
fig, ax = plt.subplots(1,2)

ax[0].hist(v_rest_sst, color='purple')
ax[0].set_ylabel('sst Resting Membrane Potential')
ax[0].set_xlabel('Count')

if pval > 0.05:
  ax[1].hist(v_rest_pvalb)
  ax[1].set_ylabel('pvalb Resting Membrane Potential')
  ax[1].set_xlabel9('Count')

```

```python
#Defines a function to perform Step 4

def vrestHist():
  print(np.mean(v_rest_sst))
  print(np.mean(v_rest_pvalb))

  t_sat, pval = (stats.ttest_ind(v_rest_sst, v_rest_pvalb))
  print(pval)

  fig, ax = plt.subplots(1,2)

  ax[0].hist(v_rest_sst, color='purple')
  ax[0].set_ylabel('sst Resting Membrane Potential')
  ax[0].set_xlabel('Count')

  if pval > 0.05:
    ax[1].hist(v_rest_pvalb)
    ax[1].set_ylabel('pvalb Resting Membrane Potential')
    ax[1].set_xlabel9('Count')

```

```python
#Calls vrestHist

vrestHist()
```

```python
#Loops through v_rest data to make a scatter plot with sst on x-axis and pvalb on y-axis

vrest_length = len(v_rest_pvalb) #both are 500 so this can just be one variable

for i in range(vrest_length):
  if v_rest_sst[i] > -60 and v_rest_pvalb[i] > -60:
    plt.scatter(v_rest_sst[i], v_rest_pvalb[i], color='red')
  if v_rest_sst[i] > -60 and v_rest_pvalb[i] < -60:
    plt.scatter(v_rest_sst[i], v_rest_pvalb[i], color='green')
  if v_rest_sst[i] < -60 and v_rest_pvalb[i] < -60:
    plt.scatter(v_rest_sst[i], v_rest_pvalb[i], color='blue')
  if v_rest_sst[i] < -60 and v_rest_pvalb[i] > -60:
    plt.scatter(v_rest_sst[i], v_rest_pvalb[i], color='purple')

plt.ylabel('Pvalb Membrane Potential')
plt.xlabel('Sst Membrane Potential')
```
